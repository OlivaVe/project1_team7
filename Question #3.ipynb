# Import libraries.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
import numpy as np
from matplotlib import pyplot as plt


df = pd.read_csv('1900_2021_DISASTER.csv')
```

# Show the correlation

```python
sns.heatmap(df.corr())
plt.savefig('HeatMapCorr.png')
plt.show()
```

```python
df_DS=df[['Disaster Subgroup','Disaster Type']]
df_DS_F=df_DS[df_DS['Disaster Subgroup']=='Geophysical']
df_DS_F
```

```python
df_DS_F['Disaster Type'].unique()
```

```python
df_DS_F['Disaster Type'].value_counts()
```

```python
df_DS_F['Disaster Type'].value_counts().plot(kind='bar' , x='Disaster Type',alpha=0.35, rot=0, color='Blue')
plt.savefig('Geophysical.png')
plt.show()
```

```python
df_DS=df[['Disaster Subgroup','Disaster Type']]
df_DS_F=df_DS[df_DS['Disaster Subgroup']=='Hydrological']
df_DS_F['Disaster Type'].value_counts().plot(kind='bar' , x='Disaster Type',alpha=0.35, rot=0, color='Blue')
plt.savefig('Hydrological.png')
plt.show()
```
