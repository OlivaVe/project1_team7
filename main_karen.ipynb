```python
import pandas as pd 
import numpy as np
import plotly.express as px 
import matplotlib.pyplot as plt
from matplotlib import pyplot as plt

main_data = pd.read_csv('./data/disasters.csv')
```

```python
hydrological = main_data[main_data["Disaster Subgroup"] == 'Hydrological']
hydrological_2000 = hydrological[hydrological["Year"] > 2000]
```

```python
map_fig = px.scatter_geo(hydrological_2000,
                         locations = 'ISO',
                         projection = 'orthographic',
                         color = 'Disaster Type',
                         opacity = 0.8,
                         hover_name = 'Country',
                         hover_data = ['Country','Continent'],
                         title = 'Disaster Type from 2000 to present'

)
map_fig.show()
map_fig.write_image("disastertype2000.png")
```

```python
drought = main_data[main_data["Disaster Type"] == 'Drought']
drought_2000 = drought[drought["Year"] > 2000]
```

```python
map_fig = px.scatter_geo(drought_2000,
                         locations = 'ISO',
                         projection = 'orthographic',
                         color = 'Disaster Type',
                         opacity = 0.8,
                         hover_name = 'Country',
                         hover_data = ['Country','Continent'],
                         title = 'Droughts from 2000 to present'

)

map_fig.show()
map_fig.write_image("drought2000.png")
```
